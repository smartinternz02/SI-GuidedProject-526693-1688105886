### **Importing libraries**

```python
import cv2
import numpy as np
import easyocr
import matplotlib.pyplot as plt
```

### **Taking Input Image**

```python
img = cv2.imread('/N201.jpeg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### **Importing our pre trained model**

```python
haarCas = '/haarcascade_russian_plate_number.xml'
haarCas
```

```python
NumberP = cv2.CascadeClassifier(haarCas)
NumberP
```

### **Finding Coordinates of image**

```python
npp = NumberP.detectMultiScale(img,scaleFactor=1.1,minNeighbors=4)
npp
```

### **Displaying number plate region**





```python
for (x,y,w,h) in npp:
   cropimg=cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),7)

plt.imshow(cropimg)
```

### **Image to text conversion**

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(cropimg)
result
```
